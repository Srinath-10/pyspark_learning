```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("learning").getOrCreate()

data = [("1", "101", "Laptop", 1200), ("2", "102", "Headphones", 150),
        ("3", "101", "Mouse", 25), ("4", "103", "Laptop", 1200),
        ("5", "104", "Monitor", 300)]
columns = ["Order_ID", "Customer_ID", "Product", "Price"]

df=spark.createDataFrame(data,schema=columns)
df.show()
```
